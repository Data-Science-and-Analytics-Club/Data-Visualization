**Data Visualization Techniques using matplotlib and seaborn**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('cars.csv')

# Display the first few rows of the dataset
print(df.head())
```

**Barcharts:**
A visual representation of data using rectangular bars of varying lengths to show the quantities or values associated with different categories.

```python
import matplotlib.pyplot as plt

# Sample data
subject = ['Statistics', 'AOA', 'Discrete Math','Data Structure']
values = [30, 45, 25, 60]

# Creating a bar chart
plt.bar(subject, values)
plt.xlabel('Electives')
plt.ylabel('No. of students')
plt.title('Bar Chart')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['fuel'].value_counts().plot(kind='bar', color='red')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.title('Bar Chart - Fuel Types')
plt.show()
```

**Line plots:** A graph that displays data points connected by straight lines, commonly used to depict trends or patterns over a continuous interval.

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 7]

# Creating a line plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='selling_price', data=df)
plt.xlabel('Date')
plt.ylabel('Selling Price')
plt.title('Line Plot - Selling Price over Time')
plt.show()
```

**Scatter plots:** A graphical representation of individual data points on a two-dimensional plane, used to explore relationships between two continuous variables.

```python
# Sample data
x = [1, 2, 3, 4, 5, 8, 9, 6, 7]
y = [10, 12, 5, 8, 7, 2, 3, 4, 13]

# Creating a scatter plot
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='km_driven', y='selling_price', data=df)
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price')
plt.title('Scatter Plot - Selling Price vs. Kilometers Driven')
plt.show()
```

**Histograms:** A graphical representation of the distribution of a dataset, depicting the frequency of data points within specified intervals, typically represented as bars.

```python
# Sample data
data = [2, 3, 3, 4, 4, 4, 5, 5, 5, 5, 6, 6, 6, 6, 6]

# Creating a histogram
plt.hist(data, bins=10, edgecolor='red')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['year'], bins=20, edgecolor='black', color='orange')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Histogram - Year Distribution')
plt.show()
```

**Box plots:** A statistical visualization that displays the distribution of a dataset, including its median, quartiles, and potential outliers, using a rectangular box and whiskers.

```python
import seaborn as sns

# Sample data
data = [30, 45, 25, 50, 60, 75, 80]

# Creating a box plot
sns.boxplot(x=data)
plt.xlabel('Values')
plt.title('Box Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='selling_price', data=df)
plt.xlabel('Selling Price')
plt.title('Box Plot - Selling Price Distribution')
plt.show()
```

**Pie charts:** A circular statistical graphic divided into slices to illustrate numerical proportions, with each slice representing a relative portion of the whole.

```python
# Sample data
sizes = [45, 35, 20]

# Creating a pie chart
plt.pie(sizes, labels=['Literate', 'Illiterate', 'Undocumented'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['seller_type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen'])
plt.title('Pie Chart - Seller Type Distribution')
plt.show()
```

**Heatmaps:** A graphical representation of data where values in a matrix are represented as colors, often used to visualize relationships or patterns in large datasets.

```python
import seaborn as sns
import numpy as np

# Sample data
data = np.random.rand(10, 10)

# Creating a heatmap
sns.heatmap(data, cmap='viridis')
plt.title('Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap - Correlation Matrix')
plt.show()
```

**Area charts:** A graph depicting quantitative data with the area below the line filled with color, providing a visual representation of cumulative totals over time or categories.

```python
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 12, 5, 8, 7]
y2 = [2, 5, 8, 11, 14]

# Creating an area chart
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Chart')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.sort_values(by='Date', inplace=True)
plt.fill_between(df['Date'], df['selling_price'].cumsum(), color='skyblue', alpha=0.4)
plt.xlabel('Date')
plt.ylabel('Cumulative Selling Price')
plt.title('Area Chart - Cumulative Selling Price over Time')
plt.show()
```

**Violin plots:** A method of plotting numeric data that combines aspects of box plots and kernel density plots to show the distribution of the data along with its probability density.

```python
# Sample data
data = [30, 45, 25, 50, 60, 75, 80]

# Creating a violin plot
sns.violinplot(x=data)
plt.xlabel('Values')
plt.title('Violin Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='transmission', y='selling_price', data=df)
plt.xlabel('Transmission')
plt.ylabel('Selling Price')
plt.title('Violin Plot - Transmission vs. Selling Price')
plt.show()
```

**Bubble charts:** A scatter plot in which a third dimension of the data is shown through the size of markers, typically representing a third variable in addition to the two plotted on the x and y axes.

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 7]
sizes = [20, 50, 10, 30, 40]

# Creating a bubble chart
plt.scatter(x, y, s=sizes)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bubble Chart')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='km_driven', y='selling_price', hue='fuel', size='fuel', data=df)
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price')
plt.title('Bubble Chart - Kilometers Driven vs. Selling Price with Fuel Type')
plt.show()
```
